#PROYECTO - EMPRESA DE DISTRIBUCIÓN DE CABLES

Esta empresa de distribución de cables, está situada en Bogotá, y se especializa en el transporte de tres diferentes tipos de cables eléctricos, cables de potencia, de goma, de aluminio, los cuales son distribuidos a diferentes ciudades de Colombia.

La empresa toman cuenta al momento del cobro, el precio de cada cable, la longitud pedida, y la distancia que se recorre desde la capital del país. Como por ejemplo una ciudad cercana a Bogotá como lo es Villavicencio, tendría un precio menor a una ciudad lejana como en caso de Cartagena.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
cables = pd.read_excel("proyecto.xlsx")

cables
```

En la tabla anterior podemos apreciar el envío que se realizará a 12 diferentes ciudades del país, teniendo los datos del tipo de cable el precio total, la longitud del cable requerida en metros, y la distancia a recorrer en kilómetros.

```python
sns.set_theme(style="whitegrid", context="talk")
sns.relplot(data=cables,
           x="Ciudad",
           y="Precio ($)",
            kind ="line",
            height = 7,
            aspect=3)
plt.show()
```

```python
sns.set_theme(style="whitegrid", context="talk")
sns.relplot(data=cables,
           x="Ciudad",
           y="Longitud (m)",
            kind ="line",
            height = 7,
            aspect=3)
plt.show()
```

```python
sns.set_theme(style="whitegrid", context="talk")
sns.relplot(data=cables,
           x="Ciudad",
           y="Distancia (Km)",
            kind ="line",
            height = 7,
            aspect=3)
plt.show()
```

En los gráficos anteriores, se puede notar las ciudades donde habrá un mayor precio a pagar, la ciudades que pidieron mayor cantidad de cable, y la ciudades más alejadas de Bogotá.

```python
Precio_total = cables["Precio ($)"].sum()
print("$",Precio_total,)
```
